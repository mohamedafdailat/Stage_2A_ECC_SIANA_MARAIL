**FFT des différents signaux tracés**

```python
# prompt: tracer la FFT des différents signaux tracés sur une fréquence de 200 Hz

def plot_fft(signal_values, sampling_frequency=200, title="FFT du Signal"):
  """
  Calcule et trace la Transformée de Fourier Rapide (FFT) d'un signal.

  Args:
    signal_values (pd.Series or np.array): Les valeurs du signal.
    sampling_frequency (int): La fréquence d'échantillonnage du signal en Hz.
    title (str): Titre du graphique FFT.
  """
  n = len(signal_values)
  fft_values = np.fft.fft(signal_values)
  frequencies = np.fft.fftfreq(n, d=1/sampling_frequency)

  # On prend la magnitude et on se concentre sur les fréquences positives
  fft_magnitude = 2.0/n * np.abs(fft_values[0:n//2])
  frequencies_positive = frequencies[0:n//2]

  plt.figure(figsize=(12, 6))
  plt.plot(frequencies_positive, fft_magnitude)
  plt.title(title)
  plt.xlabel('Fréquence (Hz)')
  plt.ylabel('Magnitude')
  plt.grid(True)
  plt.show()

# Example usage for one of the loaded signals (replace with the actual variable name)
# Assuming `signal_values` holds the signal data from one of the loaded files
# You would call this function for each signal you loaded and want to plot the FFT for.

# Example for signal_Ball_charged10.csv (assuming signal_values is the variable)
# plot_fft(signal_values, sampling_frequency=200, title='FFT du signal défaut de bille à 10 Hz')

# You would repeat the call to plot_fft for each loaded signal:
# For signal_SAIN_charged20.csv:
# plot_fft(signal_values_sain_charged, sampling_frequency=200, title='FFT du signal sain') # assuming a variable name

# For signal_OR_charged20.csv:
# plot_fft(signal_values_or_charged, sampling_frequency=200, title='FFT du signal défaut de bague externe à 20 Hz') # assuming a variable name

# And so on for all your loaded signals, using the correct variable name for each signal.

# Here's how you might integrate it with your existing loading code.
# After successfully loading a signal and getting `signal_values` (or similar variable):

# Example integration after loading signal_Ball_charged10.csv:
file_path = "/content/signal_Ball_charged10.csv"
try:
    signal_data = pd.read_csv(file_path)
    if not signal_data.empty and signal_data.shape[1] == 1:
        signal_column_name = signal_data.columns[0]
        signal_values_ball_charged = signal_data[signal_column_name]
        plot_fft(signal_values_ball_charged, sampling_frequency=200, title='FFT du signal défaut de bille à 10 Hz')
except Exception as e:
    print(f"Could not process FFT for {file_path}: {e}")

# Example integration after loading signal_SAIN_charged20.csv:
file_path = "/content/signal_SAIN_charged20.csv"
try:
    signal_data = pd.read_csv(file_path)
    if not signal_data.empty and signal_data.shape[1] == 1:
        signal_column_name = signal_data.columns[0]
        signal_values_sain_charged = signal_data[signal_column_name]
        plot_fft(signal_values_sain_charged, sampling_frequency=200, title='FFT du signal sain')
except Exception as e:
    print(f"Could not process FFT for {file_path}: {e}")

# Example integration after loading signal_OR_charged20.csv:
file_path = "/content/signal_OR_charged20.csv"
try:
    signal_data = pd.read_csv(file_path)
    if not signal_data.empty and signal_data.shape[1] == 1:
        signal_column_name = signal_data.columns[0]
        signal_values_or_charged = signal_data[signal_column_name]
        plot_fft(signal_values_or_charged, sampling_frequency=200, title='FFT du signal défaut de bague externe à 20 Hz')
except Exception as e:
    print(f"Could not process FFT for {file_path}: {e}")


# Example integration after loading signal_IR_charged20.csv:
file_path = "/content/signal_IR_charged20.csv"
try:
    signal_data = pd.read_csv(file_path)
    if not signal_data.empty and signal_data.shape[1] == 1:
        signal_column_name = signal_data.columns[0]
        signal_values_ir_charged = signal_data[signal_column_name]
        plot_fft(signal_values_ir_charged, sampling_frequency=200, title='FFT du signal défaut de bague interne à 20 Hz')
except Exception as e:
    print(f"Could not process FFT for {file_path}: {e}")

# Example integration after loading signal_COMB_charged20.csv:
file_path = "/content/signal_COMB_charged20.csv"
try:
    signal_data = pd.read_csv(file_path)
    if not signal_data.empty and signal_data.shape[1] == 1:
        signal_column_name = signal_data.columns[0]
        signal_values_comb_charged = signal_data[signal_column_name]
        plot_fft(signal_values_comb_charged, sampling_frequency=200, title='FFT du signal défaut combiné à 20 Hz')
except Exception as e:
    print(f"Could not process FFT for {file_path}: {e}")

# Example integration after loading safe_1987_8_vide.csv:
file_path_empty = "/content/safe_1987_8_vide.csv"
try:
    signal_data_empty = pd.read_csv(file_path_empty, usecols=['acceleration_vert'])
    if not signal_data_empty.empty and 'acceleration_vert' in signal_data_empty.columns:
        signal_values_sain_empty = signal_data_empty['acceleration_vert']
        plot_fft(signal_values_sain_empty, sampling_frequency=200, title='FFT du signal sain (cas à vide) - Accélération Verticale')
except Exception as e:
    print(f"Could not process FFT for {file_path_empty}: {e}")

# Example integration after loading outer_race_197_12empty.csv:
file_path_empty = "/content/outer_race_197_12empty.csv"
try:
    signal_data_empty = pd.read_csv(file_path_empty, usecols=['acceleration_vert'])
    if not signal_data_empty.empty and 'acceleration_vert' in signal_data_empty.columns:
        signal_values_or_empty = signal_data_empty['acceleration_vert']
        plot_fft(signal_values_or_empty, sampling_frequency=200, title='FFT du signal défaut de bague externe (cas à vide) - Accélération Verticale')
except Exception as e:
    print(f"Could not process FFT for {file_path_empty}: {e}")

# Example integration after loading inner_196_12_empty.csv:
file_path_empty = "/content/inner_196_12_empty.csv"
try:
    signal_data_empty = pd.read_csv(file_path_empty, usecols=['acceleration_vert'])
    if not signal_data_empty.empty and 'acceleration_vert' in signal_data_empty.columns:
        signal_values_ir_empty = signal_data_empty['acceleration_vert']
        plot_fft(signal_values_ir_empty, sampling_frequency=200, title='FFT du signal défaut de bague interne (cas à vide) - Accélération Verticale')
except Exception as e:
    print(f"Could not process FFT for {file_path_empty}: {e}")

# Example integration after loading combined_197_11_empty.csv:
file_path_empty = "/content/combined_197_11_empty.csv"
try:
    signal_data_empty = pd.read_csv(file_path_empty, usecols=['acceleration_vert'])
    if not signal_data_empty.empty and 'acceleration_vert' in signal_data_empty.columns:
        signal_values_comb_empty = signal_data_empty['acceleration_vert']
        plot_fft(signal_values_comb_empty, sampling_frequency=200, title='FFT du signal défaut combiné (cas à vide) - Accélération Verticale')
except Exception as e:
    print(f"Could not process FFT for {file_path_empty}: {e}")

# Example integration after loading ball_197_9_empty.csv:
file_path_empty = "/content/ball_197_9_empty.csv"
try:
    signal_data_empty = pd.read_csv(file_path_empty, usecols=['acceleration_vert'])
    if not signal_data_empty.empty and 'acceleration_vert' in signal_data_empty.columns:
        signal_values_ball_empty = signal_data_empty['acceleration_vert']
        plot_fft(signal_values_ball_empty, sampling_frequency=200, title='FFT du signal défaut à bille (cas à vide) - Accélération Verticale')
except Exception as e:
    print(f"Could not process FFT for {file_path_empty}: {e}")
```

**CNN**

```python
# prompt: un modèle CNN pour prédire les classes

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score # Import necessary metrics
import numpy as np # Import numpy
import matplotlib.pyplot as plt # Import matplotlib
import seaborn as sns # Import seaborn

# Le CNN travaille mieux avec des données en 3D (batch_size, timesteps, features)
# Nos données actuelles sont en 2D (samples, features). Nous devons remodeler les données.
# Ici, chaque "caractéristique" peut être considérée comme un pas de temps.
# Nous ajoutons une dimension pour représenter le nombre de caractéristiques (1 dans ce cas).
X_train_cnn = np.expand_dims(X_train_scaled, axis=2)
X_test_cnn = np.expand_dims(X_test_scaled, axis=2)

# Convertir les labels en format one-hot encoding pour le CNN
# Use the already encoded labels (y_train_encoded, y_test_encoded)
y_train_cnn = to_categorical(y_train_encoded)
y_test_cnn = to_categorical(y_test_encoded)

# Définir les paramètres du modèle CNN
input_shape = (X_train_cnn.shape[1], X_train_cnn.shape[2]) # (nombre_de_caracteristiques, 1)
# Define num_classes based on the shape of the one-hot encoded labels
num_classes = y_train_cnn.shape[1]


# Construire le modèle CNN
cnn_model = Sequential()
cnn_model.add(Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=input_shape))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Dropout(0.25))

cnn_model.add(Conv1D(filters=128, kernel_size=3, activation='relu'))
# cnn_model.add(MaxPooling1D(pool_size=2)) # Removed this layer
cnn_model.add(Dropout(0.25))

cnn_model.add(Flatten())
cnn_model.add(Dense(100, activation='relu'))
cnn_model.add(Dropout(0.5))
cnn_model.add(Dense(num_classes, activation='softmax')) # couche de sortie avec softmax pour classification multi-classes

# Compiler le modèle
cnn_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Afficher le résumé du modèle
cnn_model.summary()

# Entraîner le modèle CNN
# Vous pouvez ajuster les epochs et le batch_size selon votre dataset
history = cnn_model.fit(X_train_cnn, y_train_cnn, epochs=20, batch_size=32, validation_split=0.2)

# Évaluer le modèle CNN sur l'ensemble de test
loss, accuracy = cnn_model.evaluate(X_test_cnn, y_test_cnn, verbose=0)
print(f"\nPrécision du modèle CNN sur l'ensemble de test : {accuracy*100:.2f}%")

# Faire des prédictions et afficher le rapport de classification
y_pred_cnn_probs = cnn_model.predict(X_test_cnn)
y_pred_cnn = np.argmax(y_pred_cnn_probs, axis=1) # Obtenir l'indice de la classe prédite
y_test_original = np.argmax(y_test_cnn, axis=1) # Obtenir l'indice de la classe réelle

# Ensure consistent data types for classification_report
y_test_original = y_test_original.astype(int)
y_pred_cnn = y_pred_cnn.astype(int)

print("\nRapport de classification du modèle CNN :")
print(classification_report(y_test_original, y_pred_cnn, target_names=le.classes_))

# Matrice de confusion pour le CNN
cm_cnn = confusion_matrix(y_test_original, y_pred_cnn)
plt.figure(figsize=(10, 7))
sns.heatmap(cm_cnn, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Prédit')
plt.ylabel('Vrai')
plt.title('Matrice de confusion (CNN)')
plt.show()
```
